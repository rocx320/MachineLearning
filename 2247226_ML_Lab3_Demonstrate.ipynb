#**Mobile price dataset**

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('test.csv')
```

```python
data.head()
```

```python
data.shape
```

**Display the dataset before and after Preprocessing**

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
# min-max normalization

from sklearn.preprocessing import MinMaxScaler

# Create an instance of MinMaxScaler
scaler = MinMaxScaler()

# Perform min-max normalization
normalized_dataset1 = scaler.fit_transform(data)

normalized_dataset1
```

**Display Covariance matrix**

```python
covariance_matrix=np.cov(data)
covariance_matrix
```

**Display Eigen Values and Eigen Vector**

```python
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
```

```python
eigenvalues
```

```python
eigenvectors
```

**Display the transformed data**

```python
data = pd.DataFrame(normalized_dataset1)
data
```

#**Stroke data**

```python
df = pd.read_csv('stroke_data.csv')
```

```python
df.head()
```

```python
df.shape
```

**Display the dataset before and after Preprocessing**

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
#df['gender'].unique()
df['gender'] = df['gender'].map({"Male":1,"Female":2,"Other":3})
```

```python
#df['ever_married'].unique()
df['ever_married'] = df['ever_married'].map({"Yes":1,"No":2})
```

```python
#df['work_type'].unique()
df['work_type'] = df['work_type'].map({"Private":1,"Self-employed":2,"Govt_job":3,"children":4,"Never_worked":5})
```

```python
#df['Residence_type'].unique()
df['Residence_type'] = df['Residence_type'].map({"Urban":1,"Rural":2})
```

```python
#df['smoking_status'].unique()
df['smoking_status'] = df['smoking_status'].map({"formerly smoked":1,"never smoked":2,"smokes":3,"Unknown":4})
```

```python
# min-max normalization

from sklearn.preprocessing import MinMaxScaler

# Create an instance of MinMaxScaler
scaler = MinMaxScaler()

# Perform min-max normalization
normalized_dataset2 = scaler.fit_transform(df)

normalized_dataset2
```

**Display Covariance matrix**

```python
covariance_matrix=np.cov(df)
covariance_matrix
```

**Display Eigen Values and Eigen Vector**

```python
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
```

```python
eigenvalues
```

```python
eigenvectors
```

**Display the transformed data**

```python
df = pd.DataFrame(normalized_dataset2)
df
```
