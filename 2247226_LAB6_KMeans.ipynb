```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/data/flag.data.csv')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Load the dataset
flags_data = df

# Extract relevant features for clustering
flags_data = flags_data.select_dtypes(include=['int','float'])
```

```python
# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(flags_data)
```

```python
# Perform PCA for dimensionality reduction and visualization
pca = PCA(n_components=2)
pca_features = pca.fit_transform(scaled_features)
```

```python
# Determine the optimal number of clusters using the Elbow method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)
```

```python
# Plot the Elbow method graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
# Based on the Elbow method, let's choose the optimal number of clusters (e.g., 4)
optimal_k = 4
```

```python
# Apply K-means clustering with the chosen number of clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(scaled_features)
```

```python
# Add cluster labels to the original dataset
flags_data['Cluster'] = clusters
```

```python
# Visualize the clusters using a scatter plot
plt.figure(figsize=(10, 8))
for cluster in range(optimal_k):
    plt.scatter(
        pca_features[clusters == cluster, 0],
        pca_features[clusters == cluster, 1],
        label=f'Cluster {cluster + 1}'
    )

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-means Clustering of Flags Dataset')
plt.legend()
plt.show()
```

```python
# Load the dataset
flags_data = pd.read_csv('/content/drive/MyDrive/data/lungs.csv')  # Replace 'flags.csv' with your dataset's filename
flags_data = flags_data.select_dtypes(include=['int','float'])
flags_data = flags_data.dropna()
# Extract relevant features for clustering
# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(flags_data)

# Perform PCA for dimensionality reduction and visualization
pca = PCA(n_components=2)
pca_features = pca.fit_transform(scaled_features)

# Determine the optimal number of clusters using the Elbow method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

# Plot the Elbow method graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()

# Based on the Elbow method, let's choose the optimal number of clusters (e.g., 4)
optimal_k = 4

# Apply K-means clustering with the chosen number of clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(scaled_features)

# Add cluster labels to the original dataset
flags_data['Cluster'] = clusters

# Visualize the clusters using a scatter plot
plt.figure(figsize=(10, 8))
for cluster in range(optimal_k):
    plt.scatter(
        pca_features[clusters == cluster, 0],
        pca_features[clusters == cluster, 1],
        label=f'Cluster {cluster + 1}'
    )
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-means Clustering of Heart Dataset')
plt.legend()
plt.show()
```
